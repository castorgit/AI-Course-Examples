# MNIST relationship Time/Parameters

**Description:** Simple figure relationship parameters time<br>
**Dataset:** output of several executions <br>

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
import locale
plt.style.use('seaborn-whitegrid')
import numpy as np
import pandas as pd
```

```python
file_df = 'MNIST'
df = pd.read_csv(file_df +'.csv', header=0)
```

```python
df
```

```python

locale.setlocale(locale.LC_ALL, 'es_ES.UTF-8')
plt.rcParams['axes.formatter.use_locale'] = True
plt.rcParams["figure.figsize"] = [16,9]
plt.rcParams["figure.figsize"] = [16,9]
fig,ax = plt.subplots()
# make a plot

ax.set_xlabel('Parameters')
ax.set_ylabel('Accuracy')
ax.set_ylim([0,100])

ax.plot(df['weights'],df['Acc'], color="black", marker="o", label = "Accuracy")
ax.set_title('Accuracy vs number of neurons on MNIST dataset',color="black",fontsize=24)

fig.tight_layout()
plt.show()


```

```python
fig,ax2 = plt.subplots()
# make a plot
ax2.set_title('Time vs number of neurons on MNIST dataset',color="black",fontsize=24)

ax.set_xlabel('Parameters')
ax.set_ylabel('Time')

ax2.set_ylabel('Time')
ax2.set_ylim([0,100000])
ax2.plot(df['weights'],df['Time'], color="green", marker="o")
fig.tight_layout()
plt.show()
```
