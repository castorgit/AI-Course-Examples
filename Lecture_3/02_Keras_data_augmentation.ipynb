# Data augmentation

```python
%matplotlib inline
import matplotlib.pyplot as plt 
from numpy import expand_dims
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
```

```python
# function to plot 9 thumbnails
def plot_thumbnails(iterator_images):
    fig, axs = plt.subplots(3,3, figsize=(9,9))
    for i in range(9):
       plt.subplot(330 + 1 + i)
       batch = iterator_images.next()    # generating batch of images with iterator
       image = batch[0].astype('uint8')   # convert to unsigned integers
       plt.imshow(image)
    plt.show()
    return
```

```python
# load the image
img = load_img('./cat.jpg')
# convert to numpy array
data = img_to_array(img)
# expand dimension to one sample
samples = expand_dims(data, 0)
```

### width shift range

ImageDataGenerator(width_shift_range=[-250,250])

```python
# width shift range
datagen = ImageDataGenerator(width_shift_range=[-200,200])
# prepare iterator
it = datagen.flow(samples, batch_size=1)

plot_thumbnails(it) # plot samples to see changes in image
```

```python
# height shift range
datagen = ImageDataGenerator(height_shift_range=[-200,200])

it = datagen.flow(samples, batch_size=1)
plot_thumbnails(it) 
```

```python
# horizontal flip
datagen = ImageDataGenerator(horizontal_flip=True)
it = datagen.flow(samples, batch_size=1)
plot_thumbnails(it) 
```

```python
# vertical flip
datagen = ImageDataGenerator(vertical_flip=True)
it = datagen.flow(samples, batch_size=1)
plot_thumbnails(it) 
```

```python
# zooming image
datagen = ImageDataGenerator(zoom_range=[0.5,1.0])
it = datagen.flow(samples, batch_size=1)
plot_thumbnails(it) 
```

```python
datagen = ImageDataGenerator(brightness_range=[0.2,1.0])
it = datagen.flow(samples, batch_size=1)
plot_thumbnails(it) 
```

```python
# all together
datagen = ImageDataGenerator(width_shift_range=[-200,200],
                        height_shift_range=[-200,200],
                        vertical_flip=True,
                        horizontal_flip=True,
                        zoom_range=[0.5,1.0],
                        brightness_range=[0.2,1.0])
it = datagen.flow(samples, batch_size=1)
plot_thumbnails(it)                            
```
