# Some examples working with images

**Description:** Examples of Image management <br>
**Dataset:** Some Random images <br>

```python
%matplotlib inline
import matplotlib.pyplot as plt 
from keras.preprocessing.image import load_img, img_to_array, array_to_img
```

```python
# load the image with keras
img = load_img('./cat2.jpg')
# report details about the image
print('1 type',type(img), 'format', img.format, 'mode', img.mode, 'size', img.size)
# show the image
#img.show()
plt.imshow(img)
plt.show()
```

```python
# transform to numpy-array
print('1',type(img))
img_array = img_to_array(img)
print('2 type', img_array.dtype, 'shape',img_array.shape)
# convert back to image PIL
img_PIL = array_to_img(img_array)
print('3', type(img_PIL))
```

```python
# loading image as b&w
img_bw = load_img('./cat2.jpg', color_mode= 'grayscale')
print('1 type',type(img_bw), 'format', img_bw.format, 'mode', img_bw.mode, 'size', img_bw.size)
# show the image
img.show()
plt.imshow(img_bw, cmap='gray')
plt.show()
```
